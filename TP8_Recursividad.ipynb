## Ejercicio 1 — Factorial
Implementar `factorial(n)` de forma recursiva.

Definicion:
- Caso base: factorial(0) = 1, factorial(1) = 1
- Caso recursivo: factorial(n) = n * factorial(n-1)

Pruebas:
- factorial(5) → 120
- factorial(0) → 1
- factorial(7) → 5040

💡 **Pista:** Necesitas un caso base: ¿cuando DEJAS de recurrir? (cuando n = 0 o 1). Luego el caso recursivo multiplica n por la llamada recursiva con n-1.

```python
# Ejercicio 1: Factorial recursivo
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
print(factorial(0))
print(factorial(7))
```

## Ejercicio 2 — Fibonacci
Implementar `fibonacci(n)` de forma recursiva que retorna el n-esimo numero de Fibonacci.

Definicion:
- fib(0) = 0
- fib(1) = 1
- fib(n) = fib(n-1) + fib(n-2)

Pruebas:
- fibonacci(10) → 55
- fibonacci(6) → 8
- fibonacci(0) → 0

💡 **Pista:** Necesitas dos casos base: fib(0)=0 y fib(1)=1. El caso recursivo suma las dos llamadas previas: fib(n-1) + fib(n-2).

```python
# Ejercicio 2: Fibonacci recursivo
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(10))
print(fibonacci(6))
print(fibonacci(0))
```

## Ejercicio 3 — Potencia
Implementar `potencia(base, exponente)` de forma recursiva.

Definicion:
- Caso base: exponente = 0 → 1
- Caso recursivo: base^exp = base * potencia(base, exp-1)

Pruebas:
- potencia(2, 8) → 256
- potencia(5, 3) → 125
- potencia(10, 0) → 1

```python
# Ejercicio 3: Potencia recursiva
def potencia(base, exponente):
    if exponente == 0:
        return 1
    return base * potencia(base, exponente - 1)

print(potencia(2, 8))
print(potencia(5, 3))
print(potencia(10, 0))
```

## Ejercicio 4 — Suma de digitos
Implementar `suma_digitos(n)` que suma los digitos de un numero usando recursividad.

Pista: Usar modulo 10 y division entera.

Pruebas:
- suma_digitos(123) → 6
- suma_digitos(9999) → 36
- suma_digitos(0) → 0

```python
# Ejercicio 4: Suma de digitos
def suma_digitos(n):
    if n < 10:
        return n
    return (n % 10) + suma_digitos(n // 10)

# Pruebas
print(suma_digitos(123))
print(suma_digitos(9999))
print(suma_digitos(0))
```

## Ejercicio 5 — Busqueda en lista
Implementar `buscar_en_lista(lista, valor, indice=0)` que busca un valor recursivamente.

Retorna el indice si encuentra, -1 si no existe.

Pruebas:
- buscar_en_lista([1,2,3,4,5], 3) → 2
- buscar_en_lista([1,2,3,4,5], 10) → -1
- buscar_en_lista([], 5) → -1

```python
# Ejercicio 5: Busqueda en lista recursiva
def buscar_en_lista(lista, valor, indice=0):
    if indice == len(lista):
        return -1
    if lista[indice] == valor:
        return indice
    return buscar_en_lista(lista, valor, indice + 1)

print(buscar_en_lista([1,2,3,4,5], 3))
print(buscar_en_lista([1,2,3,4,5], 10))
print(buscar_en_lista([], 5))
```

## Ejercicio 6 — Invertir cadena
Implementar `invertir_cadena(s)` que invierte una cadena de forma recursiva.

Caso base: cadena vacia o longitud 1 → retorna igual
Caso recursivo: ultimo_caracter + invertir_cadena(resto)

Pruebas:
- invertir_cadena("hola") → "aloh"
- invertir_cadena("a") → "a"
- invertir_cadena("") → ""

```python
# Ejercicio 6: Invertir cadena recursiva
def invertir_cadena(s):
    if len(s) <= 1:
        return s
    return s[-1] + invertir_cadena(s[:-1])

print(invertir_cadena("hola"))
print(invertir_cadena("a"))
print(invertir_cadena(""))
Resultado=invertir_cadena("")
print(f'"{Resultado}"')
```

## Ejercicio 7 — Torres de Hanoi
Implementar `torres_de_hanoi(n, origen, destino, auxiliar)` que muestra los movimientos para resolver.

Reglas: Mover n discos de origen a destino usando auxiliar.

Mostrar cada movimiento como: "Mover disco de A a B"

Pruebas:
- torres_de_hanoi(2, 'A', 'C', 'B') → 3 movimientos
- torres_de_hanoi(3, 'A', 'C', 'B') → 7 movimientos

```python
# Ejercicio 7: Torres de Hanoi
def torres_de_hanoi(n, origen, destino, auxiliar):
    if n == 1:
        print(f"Mover disco de {origen} a {destino}")
        return
    torres_de_hanoi(n - 1, origen, auxiliar, destino)
    print(f"Mover disco de {origen} a {destino}")
    torres_de_hanoi(n - 1, auxiliar, destino, origen)
print("Movimientos para 2 discos:")
torres_de_hanoi(2, 'A', 'C', 'B')

print("\nMovimientos para 3 discos:")
torres_de_hanoi(3, 'A', 'C', 'B')
```

## Ejercicio 8 — MCD (Maximo Comun Divisor) - Euclides
Implementar `mcd(a, b)` usando el algoritmo de Euclides recursivamente.

Algoritmo:
- Caso base: si b = 0, retorna a
- Caso recursivo: mcd(b, a % b)

Pruebas:
- mcd(48, 18) → 6
- mcd(100, 50) → 50
- mcd(17, 19) → 1

```python
# Ejercicio 8: MCD recursivo
def mcd(a, b):
    if b == 0:
        return a
    return mcd(b, a % b)
print(mcd(48, 18))
print(mcd(100, 50))
print(mcd(17, 19))
```
